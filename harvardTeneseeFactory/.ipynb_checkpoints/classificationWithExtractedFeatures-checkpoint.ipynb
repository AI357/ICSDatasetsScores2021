```python
# 1 ou 2
def validationCrossModelDuoPourcentage(arrModels,arrData):
    tabreturn = []
    arrPrediction = []
    
    for indice, model in enumerate(arrModels):
        prediction = model.predict(arrData[indice])
        predictionCertitude = model.predict_proba(arrData[indice])
        
        size=len(predictionCertitude)
        #print(size)
        arrPrediction.append((prediction,predictionCertitude))
    divisionDessous = 0
    DivisionDessus = 0
   # 
    arrResults = [None]*size
    
    
    for indice, predictionTuple in enumerate(arrPrediction):
        for i in range(0,len(predictionTuple[1])):
            if arrResults[i] is not None:
                arrResults[i][0] +=((lambda x: x[0] if x[0]>x[1] else x[1] )(predictionTuple[1][i]))*( predictionTuple[0][i]  +1)
                arrResults[i][1] +=((lambda x: x[0] if x[0]>x[1] else x[1] )(predictionTuple[1][i]))
            else:
                
                arrResults[i] = [None] *2
                arrResults[i][0] =((lambda x: x[0] if x[0]>x[1] else x[1] )(predictionTuple[1][i]))*( predictionTuple[0][i]  +1)
                arrResults[i][1] =((lambda x: x[0] if x[0]>x[1] else x[1] )(predictionTuple[1][i]))
        
    
    for tu in arrResults:
        tabreturn.append( int(np.around(tu[0]/tu[1])) -1)
    
    
    return tabreturn
```

```python
def validationCrossModelMultiPourcentage(arrModels,arrData):
    tabreturn = []
    arrPrediction = []
    
    for indice, model in enumerate(arrModels):
        prediction = model.predict(arrData[indice])
        predictionCertitude = model.predict_proba(arrData[indice])
        size=len(prediction)
        arrPrediction.append((prediction,predictionCertitude))
    divisionDessous = 0
    DivisionDessus = 0
   # 
    arrResults = [None]*size
    
    for indice, predictionTuple in enumerate(arrPrediction):
        for i in range(0,len(predictionTuple[1])):
            if arrResults[i] is not None:
                arrResults[i][predictionTuple[0][i]] = predictionTuple[1][i][predictionTuple[0][i]]
            else:
                
                arrResults[i] = [None] * 21
                arrResults[i][predictionTuple[0][i]] = predictionTuple[1][i][predictionTuple[0][i]]
                
        
    
    for tu in arrResults:
        
        max = 0
        decision = 0
      #  print(print(tu))
        for id,vote in enumerate(tu):
            if vote == None:
                continue
            #print(vote)
            if vote > max:
                max = vote
                decision = id
        #print("res",id)
        tabreturn.append(decision)
    return tabreturn
    
    
```
