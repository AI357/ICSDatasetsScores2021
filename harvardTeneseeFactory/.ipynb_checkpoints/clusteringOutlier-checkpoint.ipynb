```python
ALL = pd.read_csv("ALLtrain")
ALL = ALL.drop(['Unnamed: 0'],axis=1)
ALL = ALL.fillna(0)
ALL.columns
```

```python
list_ndiscrt = list(ALL.columns.values)
list_ndiscrt.remove('simulationRun')
list_ndiscrt.remove('faultNumber')
list_ndiscrt.remove('sample')

pca = PCA(n_components = 15)
axisY = ALL.drop(list_ndiscrt,axis=1)
principalComponents = pca.fit_transform(ALL.drop(['faultNumber','simulationRun','sample'],axis=1))
dfpac = pd.DataFrame(data = principalComponents, columns = ['Co1','Co2','Co3','Co4','Co5','Co6','Co7','Co8','Co9','Co10','Co11','Co12','Co13','Co14','Co15'])
Finaldfpac = pd.concat([dfpac,axisY.drop(['sample'],axis=1)],axis=1)
```

```python
expected = []
predicted = []
for i in range(0,21):
    res = clf.predict(Finaldfpac[(Finaldfpac['simulationRun'] == 90) & (Finaldfpac['faultNumber']==i)].drop(['simulationRun','faultNumber'],axis=1))
    predicted =  predicted+ res.tolist()
    
    expected= expected + Finaldfpac[(Finaldfpac['simulationRun'] == 90) & (Finaldfpac['faultNumber']==i)].faultNumber.tolist()
    
    #print("pour attaque : ",i,"\n  ",pd.Series(res).value_counts())

predicted=list(map(TouniformPredict,predicted))
expected=list(map(TouniformExpect,expected))
```

```python
case = 0
tabres = {}
tabresReverse = {}
for e in range (0,500*21):
    if e%500 ==0 and e!=0:
        case +=1
    #Finaldfpac[Finaldfpac['simulationRun'] == 1].iloc[e]
    if case in tabres:
        if pred[e] in tabres[case]:
            tabres[case][pred[e]] += 1
        else:
            tabres[case][pred[e]] = 1
    else:
        tabres[case] = {}
        tabres[case][pred[e]] = 1
    if pred[e] not in tabresReverse:
        tabresReverse[pred[e]] = {}
        tabresReverse[pred[e]][case] = 1
    else:
        if case not in tabresReverse[pred[e]]:
            tabresReverse[pred[e]][case] = 1
        else:
            
            tabresReverse[pred[e]][case] += 1
print("clusters par attaques")
tabres 
```
