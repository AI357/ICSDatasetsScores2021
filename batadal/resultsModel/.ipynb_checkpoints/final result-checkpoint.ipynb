```python
#il faut maintenant combiner les résultats pour éffacer les faux positifs
```

```python
OfficialResults = dfLTestLabeled[dfLTestLabeled.columns.values[-1]].values[len(dfLTestLabeled)-2069: ]
```

```python
modelsResultsPoidPour = [1,5,1,1,1,1,1,0.5,0.5,1]#[1.5,2,0.5,3,2,1,0.5,0.5,1,10]
modelsResultsPoidcontre = [0.5,3,0.5,5,1,0.5,0.5,0.5,0.5,2]
modelsResults = [ResultOutlierGlobal,ResultGlobalOutlierFenetreLowFP,ResultGlobalOutlierFenetre,ResultTimeSeries,ResultCorrModel,ResultOutlierLocalT675,ResultOutlierLocalT4,ResultOutlierLocalT3,ResultOutlierLocalT234,ResultOutlierLocalT1]
testtab = []

pasTFlags = [0 for i in range(20)]
finalResult = []
for indice in range(0,2069):
    votePour = 0
    voteContre = 0
    for ModelResultIndice in range(0,len(modelsResults)):
        if modelsResults[ModelResultIndice][indice] == 1:
            votePour += modelsResultsPoidPour[ModelResultIndice]
        else:
            voteContre += modelsResultsPoidcontre[ModelResultIndice]
    pasTFlags.append(votePour)
    
    #print(pasTFlags)
    if votePour >= 3.5:
        finalResult.append(1)
    elif np.array(pasTFlags).mean() >= 1.1  :
        finalResult.append(1)
    else :
        finalResult.append(0)
    testtab.append(votePour)
    
    
    
    pasTFlags.pop(0)
```

```python
plt.figure(figsize=(16,5))
plt.suptitle('final results',fontsize=19)
plt.plot(OfficialResults)
plt.plot(finalResult)
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(OfficialResults, finalResult))
```

```python
from sklearn.metrics import f1_score
f1_score(OfficialResults, finalResult, average='macro')
```

```python
from sklearn.metrics import precision_score
precision_score(OfficialResults, finalResult, average='micro')
```

```python
from sklearn.metrics import precision_score
precision_score(OfficialResults, finalResult, average='weighted')
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(OfficialResults, finalResult)
```
