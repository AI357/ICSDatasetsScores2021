```python
dfLTrainingTemp = dfLTraining.drop(['DATETIME'],axis=1)
dfLTrainingWindow = []
for i in range(0,len(dfLTrainingTemp)-2):
    attaqueBool = 0
    if dfLTrainingTemp.iloc[i][-1] == True:
        attaqueBool+=1
    if dfLTrainingTemp.iloc[i+1][-1] == True:
        attaqueBool+=1
    if dfLTrainingTemp.iloc[i+2][-1] == True:
        attaqueBool+=1
    dfLTrainingWindow.append(np.concatenate((dfLTrainingTemp[t4dataNames].iloc[i].values , dfLTrainingTemp[t4dataNames].iloc[i+1].values , dfLTrainingTemp[t4dataNames].iloc[i+2].values, [(lambda attaqueBool : 1 if attaqueBool >= 2 else 0)(attaqueBool)] )))
```

```python
dfLTrainingAttacktemp = dfLTrainingAttack.drop(['DATETIME'],axis=1)
dfLTrainingAttackWindow = []
for i in range(0,len(dfLTrainingAttacktemp)-2):
    attaqueBool = 0
    if dfLTrainingAttacktemp.iloc[i][-1] == True:
        attaqueBool+=1
    if dfLTrainingAttacktemp.iloc[i+1][-1] == True:
        attaqueBool+=1
    if dfLTrainingAttacktemp.iloc[i+2][-1] == True:
        attaqueBool+=1
    dfLTrainingAttackWindow.append(np.concatenate((dfLTrainingAttacktemp[t4dataNames].iloc[i].values , dfLTrainingAttacktemp[t4dataNames].iloc[i+1].values , dfLTrainingAttacktemp[t4dataNames].iloc[i+2].values, [(lambda attaqueBool : 1 if attaqueBool >= 2 else 0)(attaqueBool)] )))
    
```

```python
dfLTestLabeledtemp = dfLTestLabeled.drop(['DATETIME'],axis=1)
dfLTestLabeledWindow = []
for i in range(0,len(dfLTestLabeledtemp)-2):
    attaqueBool = 0
    if dfLTestLabeledtemp.iloc[i][-1] == True:
        attaqueBool+=1
    if dfLTestLabeledtemp.iloc[i+1][-1] == True:
        attaqueBool+=1
    if dfLTestLabeledtemp.iloc[i+2][-1] == True:
        attaqueBool+=1
    dfLTestLabeledWindow.append(np.concatenate((dfLTestLabeledtemp[t4dataNames].iloc[i].values , dfLTestLabeledtemp[t4dataNames].iloc[i+1].values , dfLTestLabeledtemp[t4dataNames].iloc[i+2].values, [(lambda attaqueBool : 1 if attaqueBool >= 2 else 0)(attaqueBool)] )))
    
```
